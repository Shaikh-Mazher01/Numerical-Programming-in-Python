```python
from bs4 import BeautifulSoup
import requests
url='https://jobs.lever.co/angellist?'
response = requests.get(url)
soup = BeautifulSoup(response.content, 'html.parser')
soup.prettify()
```

```python
job_cat=soup.find_all('span',class_='sort-by-team posting-category small-category-label department')
job_name=soup.find_all('h5')
job_location=soup.find_all('span',class_='sort-by-location posting-category small-category-label location')
job_shift=soup.find_all('span',class_='sort-by-commitment posting-category small-category-label commitment')
job_type=soup.find_all('span',class_='display-inline-block small-category-label workplaceTypes')
```

```python
category=[]
name=[]
location=[]
j_type=[]
shift=[]
```

```python
for i in job_cat:
  category.append(i.text)
for i in job_name:
  name.append(i.text)
for i in job_location:
  location.append(i.text)
for i in job_type:
  j_type.append(i.text)
for i in job_shift:
  shift.append(i.text)
```

```python
import pandas as pd
```

```python
d={'Job Category':category,'Job Name':name,'Job Location':location,'Job Shift':shift,'Job Type':j_type}
df=pd.DataFrame(d)
```

```python
# Salaries is not given on AngelList.com therefore I using random custom salary column
from random import randint, randrange
salary=[]
for i in range(15):
  salary.append(randint(30000,400000))
```

```python
len(salary)
```

```python
df=pd.DataFrame(d)
```

```python
df['Job Salary']=salary
```
